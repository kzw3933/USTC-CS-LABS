```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [3,6,9,12,15,18]
y_count = np.array([0.000543,0.000396,0.000578,0.000698,0.001704,0.004388])
y_heap = np.array([0.000003,0.000024,0.000258,0.002712,0.023700,0.089720])
y_merge = np.array([0.000014,0.000164,0.001435,0.010269,0.048371,0.273792])
y_quick = np.array([0.000001,0.000010,0.000139,0.001036,0.014402,0.048072])

x_ticks = x
plt.xticks(x_ticks)
plt.title("comparison of 4 algorithms")
plt.plot(x,1000*y_count,'b',label="counting sort")
plt.plot(x,1000*y_heap,'m',label="heap sort")
plt.plot(x,1000*y_merge,'g',label="merge sort")
plt.plot(x,1000*y_quick,'y',label="quick sort")
plt.xlabel("x (input scale is 2^x)")
plt.ylabel("y time cost(ms)")

plt.grid(axis='y')

plt.legend(labels=["counting sort","heap sort","merge sort","quick sort"],loc="upper left",fontsize=6)  
plt.plot()
```
