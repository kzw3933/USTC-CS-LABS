```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## 矩阵链乘
x = np.array([5, 10, 15, 20, 25])
x = x
y = np.array([0.000499805, 0.00293744, 0.00943117, 0.0197038, 0.0367684])

x_ticks = x
plt.xticks(x_ticks)
plt.title("matrix chain multiply")
plt.plot(x, 1000*y,'b')
plt.xlabel("x (matrix num)")
plt.ylabel("y (time costs/ 10^-3ms)")
plt.grid(axis='y')
plt.plot()
```

```python
## 最长公共子序列
x = np.array([10, 15, 20, 25, 30])
x = x
y = np.array([ 0.00181198, 0.00383232, 0.00682443, 0.0101437, 0.0140111])

x_ticks = x
plt.xticks(x_ticks)
plt.title("longest common subsequence")
plt.plot(x, 1000*y,'b')
plt.xlabel("x: sequence length")
plt.ylabel("y: time costs (10^-3 ms)")
plt.grid(axis='y')
plt.plot()
```

```python
## 矩阵链乘
x = np.array([5, 10, 15, 20, 25])
x = x**3
y = np.array([0.000499805, 0.00293744, 0.00943117, 0.0197038, 0.0367684])

x_ticks = x
plt.xticks(x_ticks)
plt.title("matrix chain multiply")
plt.plot(x, 1000*y,'b')
plt.xlabel("x^3 (matrix num)")
plt.ylabel("y (time costs/ 10^-3ms)")
plt.grid(axis='y')
plt.plot()
```

```python
## 最长公共子序列
x = np.array([10, 15, 20, 25, 30])
x = x**2
y = np.array([ 0.00181198, 0.00383232, 0.00682443, 0.0101437, 0.0140111])

x_ticks = x
plt.xticks(x_ticks)
plt.title("longest common subsequence")
plt.plot(x, 1000*y,'b')
plt.xlabel("x^2 (sequence length)")
plt.ylabel("y (time costs/ 10^-3 ms)")
plt.grid(axis='y')
plt.plot()
```
